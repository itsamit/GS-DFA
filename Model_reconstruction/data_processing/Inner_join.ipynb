```python
# import the library
import pandas as pd
```

```python
# read csv data
df1 = pd.read_csv('Human-GEM_genes.csv')
df2 = pd.read_csv('TPM_normalized_data.csv')
```

```python
# merge the data files
inner_join = pd.merge(df1,
                      df2,
                      on ='genes',
                      how ='inner')
inner_join
```

```python
# drop Unnamed column
column_index_to_drop = 1
df = inner_join.drop(inner_join.columns[column_index_to_drop], axis=1)
```

```python
df
```

```python
# Define lists of sample names for grouping
sample_groups = [['s1', 's3', 's7', 's8', 's9', 's12', 's13', 's14', 's18', 's20'],
                 ['s2', 's4', 's5', 's6', 's10', 's11', 's15', 's16', 's17', 's19'],
                 ['s21', 's22', 's23', 's24', 's25', 's26', 's27', 's28', 's29', 's30']]
```

```python
# Calculate average values for each group of samples
average_values = []
for group in sample_groups:
    selected_samples = df[group]
    average_values.append(selected_samples.mean(axis=1))
```

```python
# Calculate average values for each group of samples
average_values = []
for group in sample_groups:
    selected_samples = df[group]
    average_values.append(selected_samples.mean(axis=1))
```

```python
# Create a new DataFrame to save the average values
new_df = pd.DataFrame(average_values).T
new_df.columns = ['non-responder_T2D', 'responder_T2D', 'control']

print(new_df)
```
